```python
fig, ax = plt.subplots()
np.random.seed(0)
data = pd.DataFrame({
    "a": np.arange(50),
    'b': np.arange(50) + 10*np.random.randn(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.abs(np.random.randn(50)) * 100
})
ax.scatter(x='a', y='b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
# OO-style
x = np.arange(10)
fig, ax = plt.subplots(layout='tight')
ax.plot(x, x, color='orange', label='linear')
ax.plot(x, x**2, color='darkblue', label='quadratic')
ax.plot(x, x**3, color="darkred", label='cubic')
ax.set_title("Title")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.legend()
```

```python
# 2.5 Styling Artists
fig, axs = plt.subplot_mosaic([
    [1],
    [2]
], layout='tight')
axs[1].plot(np.arange(50), np.arange(50)**2, **{'color': 'darkblue', 'linewidth': 3, 'linestyle': '--'})
axs[2].plot(np.arange(50), np.arange(50), **{'color': 'darkred', 'linewidth': 1, 'linestyle': '-'})
axs[1].set_title("Styling Artists")
```

```python
# Colors
fig, ax = plt.subplots()
ax.scatter(np.arange(100), np.random.randint(0,100, 100), **{'facecolor': 'orange', 'edgecolor':'black'})
plt.tight_layout()
plt.show()
```

```python
# Linewidths
fig, axs = plt.subplot_mosaic([[1],[2],[3]])
ax1 = axs[1]
ax1.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 1})
ax1.set_title('Linewith: 1')
ax2 = axs[2]
ax2.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 3})
ax2.set_title('Linewith: 3')
ax3=axs[3]
ax3.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 5})
ax3.set_title('Linewith: 5')
plt.tight_layout()
```

```python
# LineStyles
fig, axs = plt.subplot_mosaic([[1],[2],[3]])
ax1 = axs[1]
ax1.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 1, 'linestyle':'solid'})
ax1.set_title('Linestyle: solid')
ax2 = axs[2]
ax2.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 3, 'linestyle': 'dashed'})
ax2.set_title('Linestyle: dashed')
ax3=axs[3]
ax3.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 5, 'linestyle':'dotted'})
ax3.set_title('Linestyle: dotted')
plt.tight_layout()
```

```python
# Line Color
fig, axs = plt.subplot_mosaic([[1],[2],[3]])
ax1 = axs[1]
ax1.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 1, 'linestyle':'solid', 'color':'red'})
ax1.set_title('Line Color: red')
ax2 = axs[2]
ax2.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 3, 'linestyle': 'dashed', 'color':'orange'})
ax2.set_title('Line Color: orange')
ax3=axs[3]
ax3.plot(np.arange(10), np.arange(10)**2, **{'linewidth': 5, 'linestyle':'dotted', 'color': 'blue'})
ax3.set_title('Line Color: blue')
plt.tight_layout()
```

```python
# 2.6.1 Axes labels and text
mu, sigma = 115, 15
data = mu + sigma*np.random.randn(10000)
fig, ax = plt.subplots()
ax.hist(data, bins=50, facecolor='lightblue', edgecolor='black', alpha=.75)
ax.set_xlabel('Length(cm)', color='red', fontdict={'size':14, 'weight':'bold', 'family':'Times New Roman', 'style':'italic'})
ax.set_title('Hist', color='green', fontdict={'size':24, 'weight':'bold', 'family':'Times New Roman'})
ax.set_ylabel('Count', color='orange', size=18, family='Times New Roman', weight='bold', style='italic')
plt.tight_layout()
```

```python
# Annottaion
fig, ax = plt.subplots()
ax.plot(np.arange(0,5, 0.01), np.cos(np.arange(0, 5, 0.01)*2*np.pi), color='green', linewidth=3)
ax.annotate('Max', xy=(1,1), xytext=(2,1.5),arrowprops=dict(facecolor='red', edgecolor='black', shrink=0.01))
ax.annotate("Min", xy=(0.5, -1), xytext=(1.5, -1.5), arrowprops=dict(facecolor='blue', edgecolor='black',shrink=0.01))
ax.set_ylim(-2, 2)
```

```python
# 2.6.4 Legends
fig, ax = plt.subplots()
ax.plot(np.arange(10), np.arange(10), color='red', linestyle='solid', label="Label 3")
ax.plot(np.arange(10), np.arange(10)**2, color='blue',linewidth=5, linestyle='dotted', label="Label 2")
ax.scatter(np.arange(10), np.arange(10)*10, color='orange', marker='o', s=50, label='Label 3')
ax.legend(ncols=3)
```

```python
# 2.7.1 Scales
x = np.arange(100)
y = 2 ** np.random.randint(0, 10, 100)
fig, axs = plt.subplots(1,2)
axs[0].plot(x,y)
axs[1].plot(x,y)
# Y Axis Scale
axs[1].set_yscale('log')
# 2.7.2 Tick locators and formatters
axs[0].set_xticks(np.arange(0, 101, 25), ["a", "b", "c", "d", "e"])
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Y

# 
fig, ax = plt.subplots()

# 
ax.hist(np.random.randint(0, 100, 10000), bins=50, label='Hist', color='lightblue')
ax.legend(loc='upper left')

#  Y 
add_ax = ax.twinx()

# 
add_ax.plot(np.arange(100), np.arange(100), color='C2', label='Line')
add_ax.legend(loc='upper right')

#  Y 
yticks = np.arange(0, 110, 10)  # 
add_ax.set_yticks(yticks)  # 

# 
add_ax.set_yticklabels(['Zero', 'Ten', 'Twenty', 'Thirty', 'Forty', 'Fifty', 'Sixty', 'Seventy', 'Eighty', 'Ninety', 'Hundred'], color='C2')

#  Y 
add_ax.set_ylabel('Line Value', color='C2')

# 
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
sizes = np.random.randint(50, 300, 100)

#  Figure 
fig = plt.figure(figsize=(8, 6))

# 
ax1 = fig.add_axes([0, 0, 0.95, 0.95])  # 0.90.9
s1 = ax1.scatter(
    x=x, 
    y=y, 
    c=colors,  # 
    s=sizes,  # 
    edgecolors='black',
    cmap='hot_r'
)
# 
ax1.legend(
    *s1.legend_elements(
        prop="sizes",
        markerfacecolor='none', 
        markeredgecolor='black', 
        num=8
    ),
    title='Size', 
    ncol=1, 
    bbox_to_anchor=(1.105, 1),  
    frameon=False,  # 
    facecolor='none',  # 
    title_fontsize=10,  # 
    fontsize=8,  # 
    handlelength=2,  # 
    handleheight=2.5,  # 
)

#  ax2
ax2 = fig.add_axes([0.98, 0, 0.02, 0.5])  # 0.10.5
ax2.set_title('Color', size=10, loc='left')
plt.colorbar(s1, cax=ax2, extend='both')
plt.show()
```
