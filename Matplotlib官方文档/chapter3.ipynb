```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

np.random.seed(42)
data = pd.DataFrame(np.random.rand(17, 73))  # 520
data.columns = [f"Sample {i}" for i in range(1, 74)]
data.index = [f"Species {i}" for i in range(1, 18)]

fig = plt.figure(figsize=(8, 6), layout='constrained')
ax1 = fig.add_axes([0, 0, 0.1, 1])  # ax1 
#  
linkage_matrix = linkage(data.T, method='ward')  #  data.T 
dendro = dendrogram(linkage_matrix, labels=data.columns, ax=ax1, orientation='left')  # 
labels = dendro['ivl']  # 
ax1.set_frame_on(False)  # 
ax1.grid(False)          # 
ax1.set_xticks([])       # X
ax1.margins(0,0)

#  ax2 ()
ax2 = fig.add_axes([0.175, 0, 0.7, 1])  # ax2 
# 
data_sorted = data.loc[:, labels]
data_long = data_sorted.reset_index().rename(columns={'index': 'Species'}).melt(id_vars='Species', var_name='Sample', value_name='Value')
data_long['Value'] = data_long.groupby('Sample')['Value'].transform(lambda x: x / x.sum())
# 
bottom = np.zeros(len(labels))  # 
colors = plt.cm.tab20.colors    #  tab20 
for i, category in enumerate([f"Species {i}" for i in range(1, 18)]):
    category_data = data_long.query(f"Species == '{category}'")
    ax2.barh(category_data['Sample'], category_data['Value'], left=bottom, label=category, color=colors[i])
    bottom += category_data['Value'].values  # 

# 
ax2.set_frame_on(False)  # 
ax2.grid(False)          # 
ax2.set_xticks([])       # X
ax2.set_yticks([])       # Y
ax2.legend(title='Species',bbox_to_anchor=(1, 1))  # 
ax2.margins(0,0)

# 
plt.show()
```

```python
import numpy as np
from collections import Counter

def calculate_entropy(bases):
    """  """
    total_bases = len(bases)
    base_counts = Counter(bases)
    entropy = 0
    for count in base_counts.values():
        p = count / total_bases
        entropy -= p * np.log2(p)  # 
    return entropy

# 
bases = ['A', 'A', 'A', 'A', 'T', "A", "A", "A", "A", "A",'A', 'A', 'A', 'A', 'T', "A", "A", "A"]
entropy = calculate_entropy(bases)
print(f"Entropy: {entropy:.4f}")
```
