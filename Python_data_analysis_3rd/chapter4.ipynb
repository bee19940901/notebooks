```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

# 随机生成数据
np.random.seed(42)
data = pd.DataFrame(np.random.rand(5, 20))  # 20个样本，每个样本有5类数据
data.columns = [f"Sample {i}" for i in range(1, 21)]
data.index = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']

# 对列进行层次聚类
linkage_matrix = linkage(data.T, method='ward')  # 使用 data.T 对列进行层次聚类

# 绘制聚类树
plt.figure(figsize=(10, 6))
dendro = dendrogram(linkage_matrix, labels=data.columns, orientation='left')  # 显示列名作为标签
labels=dendro['ivl']
plt.title('Hierarchical Clustering of Samples')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

# 随机生成数据
np.random.seed(42)
data = pd.DataFrame(np.random.rand(5, 20))  # 5个类别，20个样本
data.columns = [f"Sample {i}" for i in range(1, 21)]
data.index = ['Species A', 'Species B', 'Species C', 'Species D', 'Species E']

# 对列进行层次聚类
linkage_matrix = linkage(data.T, method='ward')  # 使用 data.T 对列进行层次聚类

# 创建一个包含两个子图的图形
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15, 6), constrained_layout=True)

# 绘制聚类树
dendro = dendrogram(linkage_matrix, labels=data.columns, ax=ax1, orientation='left')  # 显示列名作为标签
sorted_indices = dendro['leaves']  # 获取聚类后的顺序

# 将数据按照聚类顺序重新排序
sorted_data = data.iloc[:, sorted_indices]

# 将数据转换为百分比形式（行归一化）
data_percent = (sorted_data.T / sorted_data.sum(axis=1)).T  # 每行数据除以该行的总和，得到百分比

# 绘制堆叠柱状图
bottom = np.zeros(data_percent.shape[1])  # 初始化底部的起始位置（列数为样本数）
colors = plt.cm.tab20.colors  # 使用 tab20 颜色表
for i, category in enumerate(data_percent.index):
    ax2.barh(sorted_data.columns, data_percent.loc[category], left=bottom, label=category, color=colors[i])
    bottom += data_percent.loc[category].values  # 更新底部起始位置

# 设置图表标签和标题
ax2.set_xlabel('Percentage')
ax2.set_ylabel('Samples')
ax2.set_title('Stacked Percentage Bar Chart')
ax2.legend(title='Categories', loc='upper left', bbox_to_anchor=(1.05, 1))
ax1.set_ylim(ax2.get_ylim())

# 显示图表ax1.set_ylim(ax2.get_ylim())
plt.show()
```
