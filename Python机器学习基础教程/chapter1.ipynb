```python
# 将numpy数组转换为稀疏矩阵
import numpy as np
import scipy as sp
```

```python
#创建一个numpy数组
eye = np.eye(4)
```

```python
eye
```

```python
# 将numpy数组转换为SCR格式的scipy稀疏矩阵
csf_spars_matrix = sp.sparse.csr_matrix(eye)
print(csf_spars_matrix)
```

```python
# 将numpy数组转换为COO格式的稀疏矩阵
coo_sparse_matrix = sp.sparse.coo_matrix(eye)
print(coo_sparse_matrix)
```

```python
# pandas示例
import pandas as pd
from IPython.display import display
# 将字典转换为数据框
my_dict = {
    "Name": ["John", "Anna", "Peter", "Linda"],
    "Location": ["New York", "Paris", "Berlin", "London"],
    "Age": [24, 13, 53, 33]
}
my_df = pd.DataFrame(my_dict)
print(type(my_df))
my_df
```

```python
print("选择所有年龄大于30的行")
print("使用loc方法")
display(my_df.loc[my_df.Age > 30, :])
print("使用query方法")
display(my_df.query("Age > 30"))
print("直接使用布尔索引")
display(my_df[my_df.Age>30])
```

```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
print(type(iris_dataset))
print(iris_dataset.keys())
```

```python
print(iris_dataset["data"])
```

```python
print(iris_dataset.target_names)
```

```python
display(pd.DataFrame(iris_dataset.data))
```

```python
print(iris_dataset.target)
```

```python
# 将数据转换为训练数据和测试数据
from sklearn.model_selection import train_test_split
```

```python
X_tarin, X_test, y_train, y_test = train_test_split(
    iris_dataset.data,
    iris_dataset.target,
    random_state=0
)
```

```python
print(type(X_tarin)) # numpy 数组
```

```python
# 导入K近邻算法分类器
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 初始化分类器
knc = KNeighborsClassifier(n_neighbors=1);
```

```python
# 使用fit方法进行训练
knc.fit(X_tarin, y_train)
```

```python
# 需要预测的数据
# 必须是二维数组
X_new = np.array([
    [5, 2.9, 1.0, 0.2],
])
```

```python
print(type(X_new))
display(X_new.shape)
```

```python
# 使用训练好的模型进行预测
y_new =  knc.predict(X_new)
```

```python
# 从标签名称中查找对应关系，回去预测结果
print(iris_dataset.target_names[y_new[0]])
```

```python
# 模型评估
# 获取测集数据的预测标签
y_predict = knc.predict(X_test)
# 与测试集真实标签做比较
np.mean(y_predict == y_test)
```

```python
# 或者直接使用score方法来直接计算测试集的精确度
knc.score(X_test, y_test)
```
